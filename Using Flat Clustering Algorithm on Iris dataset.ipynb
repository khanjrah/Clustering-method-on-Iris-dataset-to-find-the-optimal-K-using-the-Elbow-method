```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
x = df.drop(['Id','Species'],axis =1)
x = x.values
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters =i,init = 'k-means++',max_iter =300,n_init = 10,random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
    
```

```python
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Within Cluster sum of squares')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', n_init=10, max_iter=300,  verbose=0, random_state=0)
```

```python
y_means = kmeans.fit_predict(x)

```

```python
y_means
```

```python
plt.scatter(x[y_means == 0,0],x[y_means == 0,1],s =100,c ='red',label = 'iris-setosa')
plt.scatter(x[y_means == 1,0],x[y_means == 1,1],s =100,c ='black',label = 'iris-versicolour')
plt.scatter(x[y_means == 2,0],x[y_means == 2,1],s =100,c ='green',label = 'iris-virginica')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')
plt.legend()
```

We were able to find the correct number of classes or clusters which are 3. So Elbow method was successful in its function.
